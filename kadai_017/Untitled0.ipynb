```python
class Human:
    def __init__(self, name: str, age: int):
        self.name = name
        self.age = age

    def check_adult(self) -> None:
        if self.age >= 20:
            print(f"{self.name} ({self.age}歳) は大人です。")
        else:
            print(f"{self.name} ({self.age}歳) は大人ではありません。")


people = [
    Human("太郎", 18),
    Human("花子", 20),
    Human("次郎", 25),
    Human("三郎", 17),
]

for person in people:
    person.check_adult()
```
