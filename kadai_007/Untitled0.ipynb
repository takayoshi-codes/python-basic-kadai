```python
# リスト（array）の作成
array = [
    "月曜日は晴れです",
    "火曜日は雨です",
    "水曜日は晴れです",
    "木曜日は晴れです",
    "金曜日は曇りです",
    "土曜日は曇りのち雨です",
    "日曜日は雷雨です"
]

# ディクショナリ（dictionary）の作成
dictionary = {
    "mon": "晴れ",
    "tue": "雨",
    "wed": "晴れ",
    "thu": "晴れ",
    "fri": "曇り",
    "sat": "曇りのち雨",
    "sun": "雷雨"
}

# 水曜日の天気予報を取り出して出力（リストから）
print("水曜日の天気予報:", array[2])

# 水曜日の天気予報を取り出して出力（ディクショナリから）
print("水曜日の天気予報:", dictionary["wed"])
```
